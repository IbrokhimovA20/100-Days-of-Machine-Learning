Today in our challenge we will see how to convert categorical into numerical as models take only numerical values. We will see what are dummy variables and Label, One Hot Encoders also.

```python
#First of all let's get rid of Na values as they will disturb with our work. As now out mission is just to understand the work of encodings we will user dropna method for this
df = df.dropna()
```

```python
#Now let's identify which columns has the categorical values, they are Sex,Embarked. 
#We can consider that Name is also categorical variable by I think it does not have any effect in predicting who survived or not.
#Let's get the dummy variables by their columns in another dataframes.
Sex = pd.get_dummies(df.Sex)
Embarked = pd.get_dummies(df.Embarked)
```

```python
#Now we got 2 dataframes with total 5 columns which we should add to our final dataframe.Let's add them
new_df = pd.concat([df,Sex,Embarked], axis=1)
```

```python
new_df
```

```python
#Now after adding the columns into the new dataframe we can see that now we don't need the actual columns such as Sex and Embarked we can drop them
new_df = new_df.drop(columns=["Sex", "Embarked"])
#As we got dummy variables and added them as new columns we should remove one from each new added columns as we can face "dummy variable trap". Deleting by one column from neww added daraframes will be enough
new_df = new_df.drop(columns=["female", "S"])
```

Now let's see Label and One Hot Encoders how they can help to convert categorical data into variable

```python
#These encoders we can get from sklearn library
#Let's import df also, frop Na values also
df = pd.read_csv("../titanic.csv")
df = df.dropna()
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
```

```python
#We will appy Label encoder into Sex column in our dataframe 
#create and object of label encoder
le = LabelEncoder()
#Here fit_transform method will take the categorical values and turn it into numerical itself
df["Sex"]=le.fit_transform(df["Sex"])
```

```python
#As you can see our Sex column has become only 0 and 1
df
```

```python
#Now let's apply One hot Encoder to our Embarked column
ohe = OneHotEncoder()
df = pd.get_dummies(df, columns=["Embarked"], drop_first=True)
```

```python
#Here you can see that our categorical column Embarked removed and we got extra 2 columns like "Embarked_Q" and "Embarked_S" which became numerical As we said that we should remove column itself
#And for not getting into "dummy variable trap" encoder itself has deleted one extra column. 
df
```
