Today I will learn what is gradient descent and implement it using python. Gradient Descent is used in order to minimize the lost.
It helps a lot for example in solving Linear Regression model.

```python
#In order to implement gradient descent I will use numpy package
#In order to plot our gradient descent we will use matplotlib

#First let's see just plot graph of our function x squared
inputs = []
outputs = []
import numpy as np
for i in range(1000):
    random_number = np.random.uniform(-10, 10)
    inputs.append(random_number)
    y = random_number**2
    outputs.append(y)
```

```python
#Here we will plot the function x square
import matplotlib.pyplot as plt
plt.scatter(inputs, outputs)
```

```python
#Now let's define what is gradient. Gradient can help you in order to minimze the loss of your model. from above plot we can imagine that we should come from up till the bottom.
#This is done by the use of derivative. Now let's try to implement it
#First let's define the function of derivative as now our function is x square our derivative will be equal to 2 * x
def derivative(x):
    return 2 * x
```

```python
#Here we will do it dirst let's define inputs than evaluations
inputs_another = []
evaluations = []
#Step size is used to track the step of our gradient
step_size = 0.1
#It is usually random trying get first X
x = np.random.uniform(-10, 10)
for steps in range(20):
    inputs_another.append(x)
    evaluations.append(x**2)

    gradient = derivative(x)
    new_x = x - step_size * gradient
    x = new_x
```

```python
plt.scatter(inputs, outputs)
plt.scatter(inputs_another, evaluations, color = 'red')
```
