And also we will plot boxplots in order to see the outliers.

```python
#Z score formula is Z_score = (x-mean)/std. It shows us how the point is far away from mean or not
#So we can see that if the value we achieve is more than a threshold. Threshold lots of time is 3.
#As we consider more or less than 3 standard deviation
#First af all let's go one by one for each column and calculate every number in our columns
threshold = 3
for column in df.columns:
    number_of_outliers = 0
    column_mean = df[column].mean()
    column_std = df[column].std()
    for val in df[column].values:
        z_score = (val - column_mean) / column_std
        if z_score > threshold:
            number_of_outliers += 1
    print(column, "Length of outliers ", number_of_outliers)
```

We had printed the number of outliers each columns has using the Z score we can see that "Chas" column has the most 35 outliers

```python
#Now we will se how to make boxplots in order to visually see the number of outliers
import matplotlib.pyplot as plt

for column in df.columns:
    plt.figure()
    df.boxplot([column])
```

We have plotted the box plots in order to see the outliers in our dataset and we can see that we got some datas (outliers)
which we couldn't see while using the Z score
