Today we will see Feature Engineering. This is called when you create new features from existed ones. And new variables can effect 
to increase the accuracy eng prediction of the model 

```python
#First let's drop the Na values
df = df.dropna()
#As in our dataset we can see that there is SibSp column which means siblings/spouses aboard and Parch parents/children aboard
#So if we add all of them we can have the full family size
df['FamilySize'] = df['SibSp'] + df['Parch'] + 1  # Add 1 for the passenger themselves
```

```python
#Now in our dataset we can see that we created a new feature like FamilySize
df
```

```python
#Now from created new feature we can create a new feature while checking is the person is alone 
df['IsAlone'] = (df['FamilySize'] == 1).astype(int)
```

```python
#Now we can see in our datset another new feature IsAlone
df
```

```python
#Now if see the correlation we can see that isAlone feature effects more than Parch or SibSp for example
#that means we have done good in case of creating new features 
df_numeric = df.select_dtypes(include=['float64', 'int64'])
df_numeric.corr()
```

```python
#Let's make a heatmap which clearly shows the correlation between features in our dataset
import seaborn as sb
dataplot = sb.heatmap(df_numeric.corr(), cmap="YlGnBu", annot=True)
```
