In many datasets, different features can have vastly different ranges. For example:  
- **Age** could range from 1 to 100.  
- **Income** might range from 10,000 to 1,000,000. 

If these features are used without scaling, models that rely on distance (e.g., k-Nearest Neighbors, Support Vector Machines) or gradient-based methods (e.g., Logistic Regression, Neural Networks) may give undue importance to features with larger values (like income in this example) simply because of their magnitude.

Effect on Model Training:
- **Large** values can dominate the optimization process, making it harder for the model to converge to the optimal solution. This can:
- **Lead** to longer training times.

Cause poor performance due to misaligned feature importance.

Normalization as a Solution:
Normalization scales all features to a standard range, typically [0, 1]. This ensures that:

**All features contribute equally to the model.**

**The model becomes more stable and efficient.**

```python
#For the scaler we will user MinMaxScaler. It is formula is x - min(feature) / max(feature) - min(feature). After calculating and putting every column into this formula we will get normalized data
#We will write calculating the formula by ourselves then we will use scikit learn which helps us to calculate it easily

for column in df.columns:
    #Let's go one by one in columns we will find max and min values for each column
    min_value = min(df[column])
    max_value = max(df[column])
    #Here we will write formula of minmax scaler and apply it for every row
    df[column] = df[column].apply(lambda x: (x - min_value) / (max_value - min_value))
```

```python
#And if we see now dataset values are only from 0 till 1 we performed Normalization to our dataset
df
```

```python
#The second way to scale dataset with MinMaxscaler is using the package scikit learn. With the use of it we can make it very easily
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
model=scaler.fit(df)
scaled_data=model.transform(df)
df = pd.DataFrame(scaled_data, columns=df.columns)
```

```python
#Now we can see how using the scikit learn library we can easily perform Normalization to the dataset
df
```

```python
df = pd.read_csv("../BostonHousing.csv")
```

```python
#Now we will se scaling dataset using Standard Scaler. It is formula is x - mean(feature) / std(feature).
#After calculating and putting every column into this formula we will get scaled data
#We will write calculating the formula by ourselves then we will use scikit learn which helps us to calculate it easily

for column in df.columns:
    #Let's go one by one in columns we will find max and min values for each column
    std = df[column].std()
    mean = df[column].mean()
    #Here we will write formula of standard scaler and apply it for every row
    df[column] = df[column].apply(lambda x: (x - mean) / std)
```

```python
#Now we can see scaled dataset with the use of Standard Scaler
df
```
