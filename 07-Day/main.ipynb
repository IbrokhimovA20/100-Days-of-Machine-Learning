```python
#okay now let's work with the missing values as we already know we have missing values in Age, Cabin and Embarked columns
#So let's start from the Age column. There are many ways to work with missing values for example drop all missing values or fullfill the dataframe with median or mean values or just forward or backward fill
#In this case I guess the best method will be to drop all rwos with missing Age values as if we fullfill dataset with another values it will cause misinformation in dataset 
df = df.dropna(axis=0, subset=["Age"])
#here we dropped the rows where Age is not defined values
```

```python
#Our next target is Cabin I guess it will be correct if we will drop the whole Cabin column totally as more than half values are missing in this column
df = df.drop(columns=["Cabin"])
```

```python
#Now only we left embarked column with Na values 
#As we see only 2 rows has Na values in dataset and Embraked means the port where a passenger boarded the ship so I think here we can just fill the Na values with the mode from Embraked column
#As mode means the most frequent elements from features most probably that 2 people are also from that port which is mostly represented in dataset
# Let's implement the code of it
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True) 
```

```python
#Let's check another time wheter we have missing values in our dataset or not
df.isnull().sum()
```

```python
#Now let's identify which columns has the categorical values, they are Sex,Embarked. 
#We can consider that Name is also categorical variable by I think it does not have any effect in predicting who survived or not.
#Let's get the dummy variables by their columns in another dataframes.
Sex = pd.get_dummies(df.Sex)
Embarked = pd.get_dummies(df.Embarked)
```

```python
#Now we got 2 dataframes with total 5 columns which we should add to our final dataframe.Let's add them
df = pd.concat([df,Sex,Embarked], axis=1)
```

```python
df
```

```python
#These encoders we can get from sklearn library
#Let's import df also, frop Na values also
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
```

```python
#We will appy Label encoder into Sex column in our dataframe 
#create and object of label encoder
le = LabelEncoder()
#Here fit_transform method will take the categorical values and turn it into numerical itself
df["Sex"]=le.fit_transform(df["Sex"])
```

```python
#Now let's apply One hot Encoder to our Embarked column
ohe = OneHotEncoder()
df = pd.get_dummies(df, columns=["Embarked"], drop_first=True)
```

```python
df_numeric = df.select_dtypes(include=['float64', 'int64'])
df_numeric.corr()
```

```python
import seaborn as sb
dataplot = sb.heatmap(df_numeric.corr(), cmap="YlGnBu", annot=True)
```

```python
#As in our dataset we can see that there is SibSp column which means siblings/spouses aboard and Parch parents/children aboard
#So if we add all of them we can have the full family size
df['FamilySize'] = df['SibSp'] + df['Parch'] + 1  # Add 1 for the passenger themselves
```

```python
#Now from created new feature we can create a new feature while checking is the person is alone 
df['IsAlone'] = (df['FamilySize'] == 1).astype(int)
```

```python
#Now if see the correlation we can see that isAlone feature effects more than Parch or SibSp for example
#that means we have done good in case of creating new features 
df_numeric = df.select_dtypes(include=['float64', 'int64'])
df_numeric.corr()
```

```python
dataplot = sb.heatmap(df_numeric.corr(), cmap="YlGnBu", annot=True)
```
