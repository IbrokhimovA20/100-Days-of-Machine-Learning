```python
#Let's read the dataset
import pandas as pd
df = pd.read_csv("../BostonHousing.csv")
```

Today our topic will normal equation in Linear Regression. We will try to implement formula and solve the problem using normal equation.
The formula of normal equation is w=(X.T * X)**−1 * X.T * y. Where w is the coefficient. As the whole formula is y = wx + e where e is the minimum error.
From that we can calculate w.

```python
#Let's see the dataset
#We can see that all our features contains the numerical values and we should predict "MEDV" column as it median value price
df
```

```python
#So let's use the formula
import numpy as np
#For matrix inversion I will use numpy package
training_dataset = df[["crim", "zn", "indus", "chas", "nox", "rm", "age", "dis", "rad", "tax", "ptratio", "b", "lstat"]]
test_dataset = df["medv"]
w = np.linalg.pinv(training_dataset.T @ training_dataset) @ training_dataset.T @ test_dataset
```

```python
w
```

```python
#Okay now we have w  let's compute predictions now
predictions = training_dataset.values @ w
```

```python
#okay now let's see our predicted values and actual values 
for_mse = pd.concat([pd.DataFrame(predictions), pd.DataFrame(test_dataset)], axis=1)
#We can see that there are some differences between actual and predicted values in next steps let's calclate the MSE mean square error of our predictions
for_mse
```

```python
#Here let's now calculate the mean square error
sum_difference = 0
for prediction, actual in zip(for_mse[0].values, for_mse["medv"].values):
    sum_difference += (actual - prediction)**2

mse = sum_difference / for_mse.shape[0]
```

```python
#Let's see the mse
mse
#we can see that our MSE is 24 tomorrow when we implement gradient descent for our this dataset we will see how we did it to perform well. 
```
