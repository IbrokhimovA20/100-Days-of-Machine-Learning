Today we will try to implement Linear Regression with the use of Gradient Descent. We will not use any sklearn library but try to implement it from zero.

```python
#Let's import numpy library in order to generate numbers
import numpy as np
#Let's define the values x and y 
# Adjusting the range for y to have a similar spread to x
x = np.random.randint(1, 100, 100)
y = np.random.randint(100, 1000, 100)  # Reduced range for y
#Let's scale our y
y = 10 * x + np.random.randint(0, 500, 100)
```

```python
#Let's plot the values
import matplotlib.pyplot as plt
plt.scatter(x,y)
#We are seeing the graph now 
```

```python
#Let's try implementing Linear Regression
#Our formula is kx + b
#Firstly let's implement them as random number for k 5 and for b it is 10
k = 5
b = 10
Y_predicted = k * x + b
```

```python
#Now let's draw line in our scatterplot
plt.scatter(x,y)
plt.plot(x,Y_predicted, color = 'r')
#you can see that this line is not the ebst fitting line to our dataset that why we should find the best fitting one
```

```python
#The next step let's find the MSE of our model
mse = sum((y - Y_predicted)**2) / len(y)
mse
#This is our error we should try to decrease it with the use of gradient
```

```python
#Let's try to calculate gradient 
#The formula of gradient of k and b 
#If we will compute the derivative we will get this kind of formula for both of them
gradiet_k = -2 * sum(x  * (y - Y_predicted)) / len(y)
gradient_b = -2 * sum(y - Y_predicted) / len(y)
#Now let's define new k and new b
learning_rate = 0.0001
k = k - learning_rate * gradiet_k
b = b - learning_rate * gradient_b
```

```python
Y_predicted = k * x + b
plt.scatter(x, y)
plt.plot(x, Y_predicted, color='r')
plt.show()
#Now in this example when we updated k and b with the use of gradient we can see that now red line is fitting somehow better
```

```python
#Here let's iterate the process 100 times
from IPython.display import clear_output
import time
learning_rate = 0.0001
for i in range(100):
    gradiet_k = -2 * sum(x  * (y - Y_predicted)) / len(y)
    gradient_b = -2 * sum(y - Y_predicted) / len(y)
    k = k - learning_rate * gradiet_k
    b = b - learning_rate * gradient_b
    Y_predicted = k * x + b
    plt.scatter(x, y)
    plt.plot(x, Y_predicted, color='r')
    plt.show()
    clear_output(wait=True)
    time.sleep(1)
```
