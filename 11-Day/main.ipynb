```python
#Let's see what columns has null values
df.isnull().sum()
```

```python
#First drop the column Cabin as it has many null values than we will have 2 columns with Null values
df = df.drop(columns=["Cabin"])
```

```python
#Let's define our columns first
numerical_columns = ["Pclass", "Age", "SibSp", "Parch"]
#Now let's define categorical columns
categorical_columns = ["Sex", "Embarked"]
```

```python
#Now let's build our preprocessors
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
numerical_preprocessor = Pipeline([("imputer", SimpleImputer(strategy="median")), #This part of code will impute with the median values our missing values
                                    "scaler", StandardScaler()]) #This part will scale our data with standard scaler
```

```python
from sklearn.preprocessing import OneHotEncoder
categorical_preprocessor = Pipeline(
    [("imputer", SimpleImputer(strategy="most_frequent")), #this part of code will imputer missing values with the most frequent value
     ("encoder", OneHotEncoder())] #this part of code will encode our categorical features
)
```

```python
#Now let's create the full preprocessor to all columns
#In order to make operation on the columns we want we will import ColumnTransformer in which we can define all columns
from sklearn.compose import ColumnTransformer
full_preprocessor = ColumnTransformer([
    ("numerical", numerical_preprocessor, numerical_columns),
    ("categorical", categorical_preprocessor, categorical_columns)
])
```

```python
#Now we will make full pipeline 

final_pipeline = Pipeline([
    ("preprocess", full_preprocessor)
])
```
