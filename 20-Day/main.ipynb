```python
#Let's read the dataset
import pandas as pd
df = pd.read_csv("../BostonHousing.csv")
```

```python
for column in df.columns:
    #Let's go one by one in columns we will find max and min values for each column
    min_value = min(df[column])
    max_value = max(df[column])
    #Here we will write formula of minmax scaler and apply it for every row
    df[column] = df[column].apply(lambda x: (x - min_value) / (max_value - min_value))
```

```python
for column in df.columns:
    #Let's go one by one in columns we will find max and min values for each column
    std = df[column].std()
    mean = df[column].mean()
    #Here we will write formula of standard scaler and apply it for every row
    df[column] = df[column].apply(lambda x: (x - mean) / std)
```

```python
#Let's define our x features and our values y 
x = df[['crim', 'zn', 'indus', 'chas', 'nox', 'rm', 'age', 'dis', 'rad', 'tax', 'ptratio', 'b', 'lstat']]
y = df[['medv']]
#These are our values
```

```python
#As here also we will use gradient descent let's define our k and b values
import numpy as np
k = np.random.randint(1, 10, (x.shape[1], 1))
b = 3
```

```python
#Let's define the predicted values
y_predicted = np.zeros(x.shape[0])
for column,k_weight in zip(x.columns, range(0,14)):
    y_predicted += x[column] * k[k_weight, 0]
y_predicted += b
y_predicted = pd.DataFrame(y_predicted).rename(columns={"crim" : "medv"})
```

```python
learning_rate = 0.0002
for i in range(1000):
    y_predicted = np.dot(x, k) + b
    mse = np.sum((y - y_predicted)**2) / len(y)
    gradient_k = 2 * np.dot(x.T, (y_predicted- y)) / len(y)
    gradient_b = 2 * np.sum(y_predicted - y) / len(y)
    k = k - learning_rate * gradient_k
    b = b - learning_rate * gradient_b[0]
    if i % 100 == 0:
        print(mse)
```
